To find the number of tokens (words and punctuation symbols), use the len() function:

```python
len(text2)
```

Each text has a set that contains all the unique tokens in that text:

```python
print(sorted(set(text2)))
print()
print(len(set(text2)))
```

To measure the "lexical richness" of a text, you can divide the number of words in a text's set (the unique word count) by the total words in the text.  The result is the percentage of the entire text that the unique words represent.

```python
str((len(set(text2))/len(text2))*100)+"%"
```

You can count individual word occurrences with the count method:

```python
print("Sense word count:",text2.count("sense"))
print("Sensibility word count:",text2.count("sensibility"))
```
